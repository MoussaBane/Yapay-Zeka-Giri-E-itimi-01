### İleri toplulaştırma işlemleri(aggregate,filter,transform,apply)

###Aggregate

* Gerçek veri ile karşılaşıldığında yapılması gerekenler.

* İleri toplulaştırma işlemleri ise üstteki durumda işimize yarayacaktır

```python
df = pd.DataFrame({"gruplar":["A","B","C","A","B","C"],
                  "degisken1":[10,23,33,22,11,99],
                  "degisken2":[100,253,333,262,111,969]}, 
                 columns=["gruplar","degisken1","degisken2"],)
df
```

```python
df.groupby("gruplar")
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min", np.median, max])  #başkasından alınan değişken normal yazılırken(alındığı kütüphaneye bağlanarak ) pandasın içindeki
                                                                 #fonksiyonlar ise "" yada normal yazılabilinir
```

```python
df.groupby("gruplar").aggregate({"degisken1":"min","degisken2": "max"}) #değişkenler üzerinde farklı işelmeler yaptırmamızı sağlar
```

### Filter


>Filtelemek için fonksiyon tanımlayıp daha karmaşık koşullar için onu filter(fonksiyon) şeklinde yazarak kullanırız.

```python
df
```

```python
def filter_func(x):
    return x["degisken1"].std() > 9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

### Pivot Tablolar

>Pivot operatörü ile satırlar sütunlara dönüştürülür. Bunun için girdi tablo ifadeden, bir sütunu pivot olarak belirlemek gerekir.Pivot table, daha çok verileri gruplamak ve ortalama değerlerle birlikte görüntülemek amacıyla kullanılır. Verilerin tablo halinde görüntülenmesini sağlar.

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby("sex")[["survived"]].mean() #dataframe için iki parantezde yap
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean").unstack() #grupladı değişken secti ve aggretae içindeki şeyleri yazdık
```

```python
titanic.pivot_table("survived", index="sex", columns="class") #(değişken, index, başka değişken varmı)
```

```python
titanic.age.head(10)
```

```python
age = pd.cut(titanic["age"],[10,18,90])
age.head(10)
```

```python
titanic.pivot_table("survived", ["sex",age], "class")
```
